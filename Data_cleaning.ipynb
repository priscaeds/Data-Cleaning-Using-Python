```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_excel('call_list.xlsx')
df
```

```python
# remove duplicates first...rows 19 and 20, inplace true makes changes permanent or you assign a new name like

data=df.drop_duplicates()
data
```

```python
# to drop the not useful column

data=data.drop(columns="Not_Useful_Column")
```

```python
# to remove unwanted xters in the lastname column

data['Last_Name']=data.Last_Name.str.strip('._/')
data
```

```python
# l strip left and r strip right removes spaces and xters on thieir sides, while strip removes from both sides once
```

```python
# to clean the phone number column
# data.Phone_Number = data.Phone_Number.str.replace('[^a-zA-Z0-9]','')


data.Phone_Number = data.Phone_Number.str.replace('-','')
data.Phone_Number = data.Phone_Number.str.replace('/','')
data.Phone_Number = data.Phone_Number.str.replace('|','')
data
```

```python
data.Phone_Number = data.Phone_Number.apply(lambda x:str(x))
data.Phone_Number = data.Phone_Number.apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
data
```

```python
# remove the null values 

data.Phone_Number = data.Phone_Number.str.replace('Na--','')
data
```

```python
data.Phone_Number = data.Phone_Number.str.replace('nan--','')
data
```

```python
# to break the address coln into 3 colns

data[['Street','State','Zip_code']] = data.Address.str.split(',',expand=True)
data
```

```python
print(data.columns)
```

```python
data.Address = data.Address.fillna('')
data
```

```python
data[['Do_Not_Contact', 'Paying Customer',]] = data[['Do_Not_Contact', 'Paying Customer']].replace({'Y':'Yes', 'N':'No'})
data
```

```python
data = data.drop(columns='Address')
data
```

```python
data= data.fillna('unknown')
data
```
